```python
import cv2 as cv # 4.5.5
import numpy as np
from glob import glob
from matplotlib import pyplot as plt
import utils

assets_loaded = utils.load_test_assets()
```

## Load image

```python
img = cv.imread(assets_loaded['monsters'][4], cv.IMREAD_UNCHANGED)
# img = cv.imread(assets_loaded['monsters'][3], cv.IMREAD_UNCHANGED)
img = utils.bgra2rgba(img)

plt.imshow(img)
plt.show()
```

## Test OpenCV resize options

```python
methods = ['cv.INTER_NEAREST', 'cv.INTER_LINEAR', 'cv.INTER_CUBIC', 'cv.INTER_AREA', 'cv.INTER_LANCZOS4', 'cv.INTER_LINEAR_EXACT']
for i, method in enumerate([cv.INTER_NEAREST_EXACT,cv.INTER_NEAREST, cv.INTER_LINEAR, cv.INTER_CUBIC, cv.INTER_AREA, cv.INTER_LANCZOS4, cv.INTER_LINEAR_EXACT]):
    res = utils.resize_image(img, 4, method)
    plt.imshow(res)
    plt.title(methods[i])
    plt.show()
```

## Test duplicating columns and rows

```python
duplicate_matrix = lambda a, x, y: a.repeat(x, 0).repeat(y, 1)
scaled = duplicate_matrix(img, 4, 4)
plt.imshow(scaled)
plt.show()
```
